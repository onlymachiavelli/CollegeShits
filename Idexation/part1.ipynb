```python
import pandas as pd
```

```python
data = pd.read_csv("Computer_Science.csv")
```

```python
data["Data"]=data["Data"].str.lower()
```

```python
import re
```

```python
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

```python
data["Data"]=data["Data"].apply(remove_html_tags)
```

```python
def remove_url(text):
    pattern = re.compile('https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

```python
data["Data"]=data["Data"].apply(remove_url)
```

```python
import string
```

```python
punctuation = string.punctuation
punctuation
```

```python
def remove_punc(text):
    for char in punctuation:
        text=text.replace(char,'')
    return text
```

```python
data["Data"]=data["Data"].apply(remove_punc)
```

```python
from nltk import  word_tokenize
```

```python
data['tokens'] = data['Data'].apply(word_tokenize) 
```

```python

data['tokens']
```

```python
#install nltk 

import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words("english")
```

```python
def remove_stopwords(tokens):
    new_tokens = []

    for token in tokens:
        if token not in stopwords.words('english'):
            new_tokens.append(token)

    return new_tokens
```

```python
data["clean_words"]=data["tokens"].apply(remove_stopwords)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def apply_lemmatization(tokens):
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return lemmatized_tokens
```

```python
data["lemmatized_data"] = data["clean_words"].apply(apply_lemmatization)
```

```python
data
```

```python
for index, row in data.iterrows() :
    document = row["Document"]
    tokens = " ".join(row["lemmatized_data"])
    print(f"Document {document} : { {tokens} }")
```

```python
#for each token show documents 
from collections import defaultdict

inverted_index = defaultdict(list)

for index, row in data.iterrows() :
    document = row["Document"]
    for token in row["lemmatized_data"]:
        inverted_index[token].append(document)

for token, documents in inverted_index.items():
    print(f"{token} : {documents}")
```

```python
def term_frequency(term, document):
    count = 0
    for token in data.loc[data["Document"] == document]["lemmatized_data"].values[0]:
        if token == term:
            count += 1
    return count

#display the term frequency for each term in each document
for index, row in data.iterrows() :
    document = row["Document"]
    for token in row["lemmatized_data"]:
        print(f"TF({token},{document}) = {term_frequency(token, document)}")
        
```

```python

import math

def inverse_document_frequency(term):
    count = len(inverted_index[term])
    return math.log(len(data) / count)

#display the inverse document frequency for each term
for token in inverted_index:
    print(f"IDF({token}) = {inverse_document_frequency(token)}")


```

```python
def tf_idf(term, document):
    return term_frequency(term, document) * inverse_document_frequency(term)

tfidf = defaultdict(dict)

for index, row in data.iterrows() :
    document = row["Document"]
    for token in row["lemmatized_data"]:
        tfidf[document][token] = tf_idf(token, document)
```

```python

#display the tfidf values for each document
print("TFIDF values for each document")


for document, terms in tfidf.items():
    print(f"Document {document} : {terms}")

```

```python




def recherche_booleenne(collection, requete):
    resultats = []
    for document in collection:
        if all(mot in document for mot in requete):
            resultats.append(document)
    return resultats
```

```python
#vectorial model 

from sklearn.feature_extraction.text import CountVectorizer
#CountVectorizer class from the scikit-learn library, which converts a collection of text documents to a matrix of token counts.
from sklearn.metrics.pairwise import cosine_similarity
# Imports the cosine_similarity function from scikit-learn, which computes the cosine similarity between vectors.
def recherche_vectorielle(collection, requete):
    vectorizer = CountVectorizer()
    #Creates an instance of the CountVectorizer class, which will be used to convert the collection and query into a matrix of token counts.
    X = vectorizer.fit_transform(collection + [requete])
    sim_matrix = cosine_similarity(X)
    similarites = sim_matrix[:-1, -1]  
    resultats = [collection[i] for i in similarites.argsort()[::-1]]  
    return resultats

```

```python


#test the boolean search
collection = data["Data"].values
requete = ["computer", "science"]
resultats = recherche_booleenne(collection, requete)
print("Boolean search results")
for resultat in resultats:
    print(resultat)

```

```python
requete = "computer science"
resultats = recherche_vectorielle(collection, requete)
print("Fuzzy search results")
for resultat in resultats:
    print(resultat)
```
